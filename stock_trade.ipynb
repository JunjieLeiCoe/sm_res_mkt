```r
symbolData <- lapply(trade_code,function(x){
     y <- as.data.frame(get(x))
     colnames(y) <- c("open","high","low","close","volume","adjusted")
     y$date <- rownames(y)
     y$symbol <- x
     y
})
```

```r
combinedData <- do.call(rbind,symbolData)
rownames(combinedData) <- 1:nrow(combinedData)
```
